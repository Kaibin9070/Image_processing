1.請撰寫一個程式可輸入三個人的姓名與電子郵件，之後可藉由姓名查詢電子郵件

```python

a = []


for i in range(3):
    name = input("姓名: ")
    email = input("電子郵件: ")
    a.append((name, email))


search_name = input("要查的姓名: ")
found = False


for name, email in a:
    if name == search_name:
        print(f"{name} 的電子郵件是: {email}")
        found = True
        break

if not found:
    print(f"找不到名為 {search_name} 的人")
```

2.請設計一個程式輸入2維平面某個點的座標(x,y)，輸出該點所在象限。

```python
import numpy as np

x = float(input("x: "))
y = float(input("y: "))

point = np.array([x, y])

if np.all(point == 0):
    print("原點")
elif x > 0 and y > 0:
    print("第一象限")
elif x < 0 and y > 0:
    print("第二象限")
elif x < 0 and y < 0:
    print("第三象限")
elif x > 0 and y < 0:
    print("第四象限")
```

4.利用While迴圈輸出1~100所有數的和
(提示：請設定2個變數，一個作為迴圈控制及儲存當前數字，另一作為疊加總和之用)

```python

num = 1
total = 0

while num <= 100:
    total += num
    num += 1

print("1 到 100 :", total)

```

5.一份數學考卷共有14題，為了使及格率提高，給分方式如下：答對1～8題，每題以8分計；答對9題以上者，前8題仍以8分計，但第9題開始則每題以6分計。請根據以上敘述撰寫一支程式，讓使用者輸入答對題數，再經程式計算後輸出得分。
(提示：此題有雙層if...else判斷，請同學動動腦，別忘了做例外處理)

```python



a = int(input("請輸入答對的題數："))


b = 0
if a >= 1 and a <= 8:
    b = a * 8
elif a >= 9:
    b = 8 * 8 + (a - 8) * 6

if a < 1 or a > 14:
    print("答題數不在有效範圍內")
else:
    print("得分為：", b)

```

6.自訂一個函式檢查密碼是否安全，密碼須符合以下原則:

密碼只有數字或英文字且長度<6：「不安全的密碼」<br>
密碼只有數字或英文字且長度>6：「可能是安全的密碼」<br>
密碼包含英文與數字且長度>6：「安全的密碼」<br>
密碼包含英文與數字且長度>10：「非常安全的密碼」

```python
def b(a):
    長度 = len(a)

    if a.isalnum() and 長度 > 6:
        return "安全的密碼" if 長度 <= 10 else "非常安全的密碼"
    else:
        return "不安全的密碼"

c = input("密碼：")
a = b(c)
print(a)

```
